```python
df1 = pd.read_csv(file_path, )
df2 = pd.read_csv(file_path2, )
df1 =df1.drop(columns = ['name'])
df1.pop(df1.columns[0])
df1= df1.loc[:,df1.notna().all(axis=0)]
df2 =df2.drop(columns = ['name'])
df2.pop(df2.columns[0])
df2= df2.loc[:,df2.notna().all(axis=0)]

 
df1['type'] = 0
df2['type'] = 1
df3 = pd.concat([df1, df2], ignore_index=True)
df3= df3.loc[:,df3.notna().all(axis=0)]

X = df3.loc[:, df3.columns != "type"]
y = df3['type']

 
```

```python
def evaluate_model(model, x_test, y_test):
    from sklearn import metrics

    # Predict Test Data 
    y_pred = model.predict(x_test)

    # Calculate accuracy, precision, recall, f1-score, and kappa score
    acc = metrics.accuracy_score(y_test, y_pred)
    prec = metrics.precision_score(y_test, y_pred)
    rec = metrics.recall_score(y_test, y_pred)
    f1 = metrics.f1_score(y_test, y_pred)
    kappa = metrics.cohen_kappa_score(y_test, y_pred)

    # Calculate area under curve (AUC)
    y_pred_proba = model.predict_proba(x_test)[::,1]
    fpr, tpr, _ = metrics.roc_curve(y_test, y_pred_proba)
    auc = metrics.roc_auc_score(y_test, y_pred_proba)

    # Display confussion matrix
    cm = metrics.confusion_matrix(y_test, y_pred)

    return {'acc': acc, 'prec': prec, 'rec': rec, 'f1': f1, 'kappa': kappa, 
            'fpr': fpr, 'tpr': tpr, 'auc': auc, 'cm': cm}
```

```python
from sklearn.model_selection import train_test_split
target = df3['type']
X_train, X_test, y_train, y_test = train_test_split(df3 , target, 
                                                    shuffle = True, 
                                                    test_size=0.2, 
                                                    random_state=1)
```

```python
from sklearn import tree 
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier 
from sklearn.naive_bayes import GaussianNB

# Building Decision Tree model 
dtc = tree.DecisionTreeClassifier(random_state=0)
dtc.fit(X_train, y_train)
rf = RandomForestClassifier(random_state=0)
rf.fit(X_train, y_train)

knn = KNeighborsClassifier(n_neighbors = 5)
knn.fit(X_train, y_train)

nb = GaussianNB()
nb.fit(X_train, y_train)

# Evaluate Model
dtc_eval = evaluate_model(dtc, X_test, y_test)
rf_eval = evaluate_model(rf, X_test, y_test)
knn_eval = evaluate_model(knn, X_test, y_test)
nb_eval = evaluate_model(knn, X_test, y_test)

models = [ {'tree': dtc}, {'randomforest': rf}, {'knn': knn}, {'naive bayes':
                                                               nb}]

# Print result
print('Accuracy:', dtc_eval['acc'])
print('Precision:', dtc_eval['prec'])
print('Recall:', dtc_eval['rec'])
print('F1 Score:', dtc_eval['f1'])
print('Cohens Kappa Score:', dtc_eval['kappa'])
print('Area Under Curve:', dtc_eval['auc'])
print('Confusion Matrix:\n', dtc_eval['cm'])

# Print result
print('Accuracy:', rf_eval['acc'])
print('Precision:', rf_eval['prec'])
print('Recall:', rf_eval['rec'])
print('F1 Score:', rf_eval['f1'])
print('Cohens Kappa Score:', rf_eval['kappa'])
print('Area Under Curve:', rf_eval['auc'])
print('Confusion Matrix:\n', rf_eval['cm'])

print('Accuracy:', knn_eval['acc'])
print('Precision:', knn_eval['prec'])
print('Recall:', knn_eval['rec'])
print('F1 Score:', knn_eval['f1'])
print('Cohens Kappa Score:', knn_eval['kappa'])
print('Area Under Curve:', knn_eval['auc'])
print('Confusion Matrix:\n', knn_eval['cm'])

print('Accuracy:', nb_eval['acc'])
print('Precision:', nb_eval['prec'])
print('Recall:',nb_eval['rec'])
print('F1 Score:', nb_eval['f1'])
print('Cohens Kappa Score:', nb_eval['kappa'])
print('Area Under Curve:', nb_eval['auc'])
print('Confusion Matrix:\n', nb_eval['cm'])
```

```python
# agglomerative clustering
from numpy import unique
from numpy import where

from sklearn.cluster import AgglomerativeClustering
from sklearn.cluster import AffinityPropagation
from sklearn.cluster import Birch
from sklearn.cluster import DBSCAN
from sklearn.cluster import KMeans
from sklearn.cluster import MiniBatchKMeans
from sklearn.cluster import MeanShift
from sklearn.cluster import OPTICS
from sklearn.cluster import SpectralClustering
from sklearn.mixture import GaussianMixture

from matplotlib import pyplot


# define the model
modelagg = AgglomerativeClustering(n_clusters=2)
# fit model and predict clusters



modelb = Birch(threshold=0.01, n_clusters=2)
#modela = AffinityPropagation(damping=0.9)
#modeldb = DBSCAN(eps=0.30, min_samples=9)
modelkm = KMeans(n_clusters=2)
modelmbkm = MiniBatchKMeans(n_clusters=2)
#modelmsh = MeanShift()
modeloptics = OPTICS(eps=0.8, min_samples=10)
modelsc = SpectralClustering(n_clusters=2)
modelgm = GaussianMixture(n_components=2)

models =  [ {'agglo': modelagg},
    {'birch': modelb}, 
           #{'affinity': modela}, 
    #{'dbscan': modeldb}, 
           {'kmeans': modelkm},
    {'minbatchkmeans': modelmbkm},
    #{'meanshift': modelmsh},
           {'optics': modeloptics},
    {'spectral': modelsc}, {'gaussianmixture': modelgm},
     
]
model_name = [  list(dict.keys())[0] for dict in models]
model_name
```

```python

def train_visualize(model):
    yhat = model.fit_predict(X)
    # retrieve unique clusters
    clusters = unique(yhat)
    # create scatter plot for samples from each cluster
    for cluster in clusters:
        # get row indexes for samples with this cluster
        row_ix = where(yhat == cluster)
        # create scatter of these samples
        
        pyplot.scatter(X[row_ix, c1], X[row_ix, c2],label=class_name[cluster])
    # show the plot
    pyplot.xlabel(df3.columns[c1])
    pyplot.ylabel(df3.columns[c2])
    pyplot.legend()
    pyplot.show()
    return yhat
```

```python
from sklearn.metrics import adjusted_rand_score 
from sklearn.metrics import rand_score 
from sklearn.metrics import silhouette_score 
from sklearn.metrics import davies_bouldin_score 
from sklearn.metrics import mutual_info_score , calinski_harabasz_score

def performance_cluster(target,features,labels_):
    ari = adjusted_rand_score(target, labels_)  
    ss = silhouette_score(features,labels_) 
    ris = rand_score(target, labels_) 
     
    mis = mutual_info_score(target, labels_) 
    dbs = davies_bouldin_score(features, labels_) 
    
    #cal = calinski_harabasz_score(features, labels_)
    
    scores = {'mutual_info': mis, 'davis_bouldin': dbs, 'adjusted_rand_score':ari, 'silhouette':ss,'rand_score': ris, }# 'calinski': cal,}
                
    
    
    return scores

     
  
 
    
```

```python
yhat_array = []
for model in models:
    k,v = list(model.items())[0]
    print(k)
    yhat = train_visualize(v)
    yhat_array.append(yhat)
    
    
        
```

```python
score_array = {}
for yhat,model in zip(yhat_array,models):
    k,v = list(model.items())[0]
     
    try: 
        score_array[k] =  performance_cluster(y,X,yhat)
        print(score_array[k])
    except:
        print('cannot measure')
```
