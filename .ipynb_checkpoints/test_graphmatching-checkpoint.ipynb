```python
class GeometricRandomWalkKernel():
    __type__ = "sim"
    @staticmethod
    def maxDegree(G):
        degree_sequence = sorted(nx.degree(G).values(), reverse=True)  # degree sequence

        # print "Degree sequence", degree_sequence
        dmax = max(degree_sequence)
        return dmax
    @staticmethod
    def compare(listgs):

        n = len(listgs)
        comparison_matrix=np.zeros((n,n))
        for i in range(n):
            for j in range(i,n):
                if len(listgs[i]) <1 or len(listgs[j]) <1:
                    comparison_matrix[i, j] = 0
                    comparison_matrix[j, i] = 0
                    continue
                direct_product_graph=nx.tensor_product(listgs[i],listgs[j])
                Ax = nx.adjacency_matrix(direct_product_graph).todense()
                try:
                    la = 1/ ((GeometricRandomWalkKernel.maxDegree(direct_product_graph)**2)+1) # lambda value
                except:
                    la= pow(1,-6)
                eps = pow(10,-10)
                I=np.identity(Ax.shape[0])
                I_vec=np.ones(Ax.shape[0])
                x=I_vec.copy()
                x_pre=np.zeros(Ax.shape[0])
                c=0

                while (np.linalg.norm(x-x_pre)) > eps:
                    if c > 100:
                        break
                    x_pre=x

                    x= I_vec + la*np.dot(Ax,x_pre.T)
                    c+=1
                comparison_matrix[i,j]=np.sum(x)
                comparison_matrix[j,i]=comparison_matrix[i,j]
        print(comparison_matrix)
        for i in range(n):
            for j in range(i,n):
                comparison_matrix[i,j] = (comparison_matrix[i,j]/np.sqrt(comparison_matrix[i,i]*comparison_matrix[j,j]))
                comparison_matrix[j,i]=comparison_matrix[i,j]
        return comparison_matrix
```

```python

# Gmatch4py use networkx graph 
import networkx as nx 
# import the GED using the munkres algorithm

import numpy as np
from matplotlib import pyplot as plt
import os
 
 
import gmatch4py as gm

num_ele = 40
```

```python
G_list = []
 

item_keys =  ['name', 'edges', 'nodes', 'clique', 'average_cluster', 'g_eff', 'triangle', 'g_reach', 'min_cycle_basis',
              'max_ind_set', 'max_matching', 'num_isolate', 's_materic', 'tree_branching_weight', 'closeness_vitality',
             #'eccentric',   'center'
              'num_clique','min_vertex_cover',
              'dominating_set','clustering','assort_coeff','pearson_coeff',
              'avg_neighbor_degree','avg_degree_connectivity',
              'num_connected_component',
             'num_connected','min_weight_matching'] # 'weiner_index']
 
item_dict = dict(zip(item_keys, ([] for name in item_keys)))
 
    

#print(item_dict)



for file2 in  output_file[:num_ele]:
     
    path,fname =  os.path.split(file2)
    fname = fname.split('.')[0]
    
    G = nx.Graph()
    g_item = dict()
    g_item['name'] = fname
    with   open(file2) as f:

        for line in f.readlines():
            l = line.strip()
            l = l.split()
            n1,n2,w = float(l[0]), float(l[1]), float(l[2])
            #print(n1,n2,w)
            G.add_edge(n1,n2,weight=w)
            
    print('fname %s edges %d nodes %d'%(fname,len(G.edges()),len(G.nodes)))
    #print('nodes %d'%(len(G.nodes())))
    item_dict['name'].append(  fname)
    item_dict['edges'].append( len( G.edges()))
    item_dict['nodes'].append(  len(G.nodes()))
    item_dict['clique'].append(  len(nx.node_clique_number(G)))
    item_dict['num_clique'].append(len(nx.algorithms.clique.number_of_cliques(g1)))
    
    item_dict['average_cluster'].append( nx.average_clustering(G))
    #item_dict['radius'] = nx.radius(G)
    item_dict['g_eff'].append(nx.global_efficiency(G))
    item_dict['triangle'].append(  sum(nx.algorithms.triangles(G)))
    item_dict['g_reach'].append(   nx.global_reaching_centrality(G))
    item_dict['min_cycle_basis'].append(   len(nx.minimum_cycle_basis(G)))
    #item_dict['cycles'] = len(nx.find_cycle(G))
    
    #item_dict['avg_shortest_path'] =  nx.average_shortest_path_length(G)
    item_dict['max_ind_set'].append(len(nx.maximal_independent_set(G)))
    item_dict['min_weight_matching'].append(len(nx.min_weight_matching(G)))
 
    item_dict['max_matching'].append(   len(nx.maximal_matching(G)))
    item_dict['num_isolate'].append(   nx.number_of_isolates(G))
    
    item_dict['s_materic'].append(   nx.s_metric(G, normalized=False))
    item_dict['tree_branching_weight'].append(    nx.tree.branching_weight(G))
    item_dict['closeness_vitality'].append(  sum( nx.closeness_vitality(G)))
     
    item_dict['num_connected'].append(nx.number_connected_components(G))
    
    item_dict['min_vertex_cover'].append(len(nx.algorithms.approximation.min_weighted_vertex_cover(G)))
    item_dict['dominating_set'].append( len(nx.dominating_set(G)))
    
    item_dict['assort_coeff'].append(nx.degree_assortativity_coefficient(G))
    item_dict['pearson_coeff'].append(nx.degree_pearson_correlation_coefficient((G)))
    
    item_dict['num_connected_component'].append(nx.number_connected_components(G))
    item_dict['clustering'].append(sum(nx.clustering(G)))
    item_dict['avg_neighbor_degree'].append(sum(nx.average_neighbor_degree(G))/len(nx.average_neighbor_degree(G)) )
    item_dict['avg_degree_connectivity'].append(sum(nx.average_degree_connectivity(G)) / len(nx.average_degree_connectivity(G)))
    #item_dict['eccentric'].append(sum(nx.eccentricity(G)))
    #item_dict['center'].append(len(nx.center(G)))
    
    #item_dict['weiner_index'].append(  nx.wiener_index(G))

    g_item['graph'] = G
     
    G_list.append(g_item)
    

```
